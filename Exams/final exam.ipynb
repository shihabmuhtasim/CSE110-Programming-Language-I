Question-1:
Write a python code that will take a string with multiple numbers separated 
by commas as an input from the user. Then create a list from the input 
string and print it. Then create a new list consisting only of the prime 
number from the previous list. Print the two lists as shown and then if the 
summation of all the elements in the new list is also a prime number then 
your program should print “It’s a miracle!!!”, otherwise your program 
should print “Just another normal day!!!”. If there is no prime number in 
the given list then keep the new list empty.
======================================================
Sample Input 1:
10, 12, 2, 17, 14

Sample Output 1:
Input list: [10, 12, 2, 17, 14]
Prime list: [2, 17]
It’s a miracle!!!

Explanation 1:
 From the input list, 2 and 17 are prime numbers. So they are added to the 
prime list. The summation of 2 and 17 is 19 which is also a prime number. 
So, “It’s a miracle!!!” is printed
=====================================================
Sample Input 2:
10, 12, 14

Sample Output 2:
Input list: [10, 12, 14]
Prime list: []
Just another normal day!!!

Explanation 2:
 From the input list, none of the numbers were prime numbers. So, the empty 
list is printed. The sum of an empty list is 0, which is not a prime 
number. So, “Just another normal day!!!” is printed.

=====================================================
Sample Input 3:
10, 12, 13, 3

Sample Output 3:
Input list: [10, 12, 13, 3]
Prime list: [13, 3]
Just another normal day!!!

Explanation 3:
 From the input list,  13 and 3 are prime numbers. The summation of 13 and 3 
is 16 which is not a prime number. So, “Just another normal day!!!” is

Question-3:
Write a function that takes two lists as arguments. If the two lists are 
not of equal length return the string “The length of the lists are not 
equal”. If the lists are equal, then take the values of the ODD index of 
List1 and multiply by 2 and use these to replace odd index values of List 
2. Then find the sum of List1 and modified List2 and return “The updated 
sum is x”. The x here is your updated sum.

======================================================
Function call 1:
function_name([5,6,7,8],[6,8,2,9])

GIVEN LISTS 1:
List1=[5,6,7,8]
List2=[6,8,2,9]

Sample Output 1:
Updated list2: [6, 12, 2, 16]
The updated sum is 62

Explanation 1:
Here, the odd index values for List1: 6*2 = 12  &  8*2 = 16
The updated List 2 will be [6,12,2,16] but the List1 will stay the same.
The updated sum for List1 and List2 then will be then 5+6+7+8+6+12+2+16 = 62
======================================================
Function call 2:
function_name([5,6,7],[6,8,2,9])

GIVEN lists 2:
List1=[5,6,7]
List2=[6,8,2,9]

Sample Output 2:
The length of the lists are not equal
======================================================


```python
#1
num_str = input("Please enter multiple numbers separated by commas:")
num_list = num_str.split(",")

for i in range(0, len(num_list)):
    num_list[i] = int(num_list[i])

prime_list = []
for i in num_list:
    flag = True
    for j in range(2, i):
        if i%j == 0:
            flag = False
    if flag == True:
        prime_list.append(i)
print("Input list:", num_list)
print("Prime list:", prime_list)

sum_primes = 0
for i in prime_list:
    sum_primes = sum_primes + i

flag = True
for i in range(2, sum_primes):
    if sum_primes % i == 0:
        flag = False
        break

if flag == True and len(prime_list) != 0:
    print("It’s a miracle!!!")
else:
    print("Just another normal day!!!")
```

```python
#3
def list_modify(li_1, li_2):
    total = 0
    if len(li_1) != len(li_2):
        return total, li_2
    else:
        for i in range(len(li_1)):
            if i % 2 != 0:
                temp = li_1[i] * 2
                li_2[i] = temp

        for i in range(len(li_1)):
            total += (li_1[i] + li_2[i])
        return total, li_2


final_sum, final_list = list_modify([5,6,7,8],[6,8,2,9])
if final_sum == 0:
    print("The length of the lists are not equal")
else:
    print("Updated list2:", final_list)
    print("The updated sum is", final_sum)
```
