### Task 1
Write a python program that will take a number as an input and check whether the number is Awesome or not. If the number is Awesome, it prints True. Otherwise False.

**Awesome number**: a number where every digit is less than its immediate left digit is called an Awesome number. A single-digit number cannot be an awesome number(e.g. 5421 is an Awesome number.)

==========================================================

**Sample Input1:**\
976321\
**Sample Output1:**\
True

==========================================================

**Sample Input2:**\
9766321\
**Sample Output2:**\
False

==========================================================

**Sample Input3:**\
9\
**Sample Output3:**\
False

==========================================================

**Sample Input4:**\
78\
**Sample Output4:**\
False

==========================================================

**Sample Input5:**\
87\
**Sample Output4:**\
True


```python
num=input("enter number:")
flag= True
if len(num)<=1:
    flag=False
elif len(num)>1:
    for i in range(len(num)-1):
        digit1= num[i]
        digit_next= num[i+1]
        if digit1<digit_next:
            flag= False
           
print(flag)           
```

### Task 5

Write a python program that will take a list of numbers(integer) as an input from the user. Finally, print the nested list to the user. 

**Sample Input1:**\
[97821, 97210, 979210]\
**Sample Output1:**\

[976321, 9640, 3211]\
**Sample Output3:**\
Data [number, sum of digits, product of digits]: [[976321, 28, 2268], [9640, 19, 0], [3211, 7, 6]]

================================================


```python
list= input("Please enter numbers seperated by comma:").split(",")
for i in range(len(list)):
    list[i]=int(list[i])
for i in range(len(list)):
    list[i]=str(list[i])
valid=[]
for i in list:
    if 3<len(i)<7:
        valid.append(int(i))
final_list=[]
for i in valid:
    sum=0
    product=1
    for dig in str(i):
        sum+=int(dig)
        product*=int(dig)
    final_list.append([i,sum,product])
print(final_list)     
print(valid)        
```
